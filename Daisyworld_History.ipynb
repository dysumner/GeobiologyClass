# Input Your Initial Conditions Here

```python
print('How many cells does your game have?')
cells = int(input())
print('What is your initial temperature?')
T = float(input())
const = 0.3
```

### The next lines initialize your Daisyworld History

```python
DW_T = [T]
DW_black = [0]
DW_white = [0]
```

## Input Your Numbers of Daisies Here

```python
play = 'y'
while play == 'y':
    empty = -1
    
    # Ask for new data
    while empty < 0:
        print('How many black daisies do you have? (Enter n to stop.)')
        ans = input()
        try: 
            int(ans)
            black = int(ans)
        except: 
            play = 'n'
            print('Not a number, so the play will stop')
        if play == 'y':
            print('How many white daisies do you have?')
            ans = input()
            try: 
                int(ans)
                white = int(ans)
            except: 
                play = 'n'
                print('Not a number, so the play will stop')

        try:
            empty = cells - black - white
        except:
            empty = 0
        if empty < 0: print("You have too many daisies! Try again. \n")
            
    if play == 'y':
        # Calculate the new temperature
        T += const*(black-white)
        print('\n The new temperature is',round(T,1), '\n\n')
   
        # Add your round to the history
        DW_T.append(T)
        DW_black.append(black)
        DW_white.append(white)
```

## The next lines plot your data:

```python
x = np.arange(0, len(DW_T))
labels = ['black','white']

fig, ax_lst = plt.subplots(2, 1)
fig.suptitle('Daisyworld History')
fig.set_size_inches(16,8)
ax_lst[1].set_xlabel('Time (in rounds played)')
ax_lst[0].set_ylabel('Temperature (Â°C)')
ax_lst[1].set_ylabel('Daisies')
ax_lst[1].set_ylim(0,cells)
ax_lst[0].plot(x, DW_T)
ax_lst[1].stackplot(x,DW_white,DW_black,labels=labels)
ax_lst[1].legend(loc='upper left')
```
